```python
import pandas as pd
import math
import matplotlib


df = pd.read_csv("Fortnite Statistics.csv")

def moyenne(liste):
    moy = sum(liste)/len(liste)
    return moy

def ecart_type(liste):
    moy = sum(liste) / len(liste)
    variance = sum((j-moy)**2 for j in liste) / len(liste)
    ecart = math.sqrt(variance)
    return str(ecart) 

def mediane(liste):
    if len(liste)%2 !=0:
        mediane = liste[(len(liste)//2 )]
    elif len(liste)%2 == 0:
        mediane = (liste[(len(liste)//2)-1]+liste[(len(liste)//2)])/2
    return mediane
def affichage(liste):
    print(f"La moyenne est: {moyenne(liste)}")
    print(f"L'écart type est: {ecart_type(liste)}")
    print(f"La médiane est : {mediane(liste)}")
    
    
```

```python
# placement normal
df = df.sort_values("Placed")
liste_placement = list(df['Placed'])
affichage(liste_placement)

#eliminations normal
print("----------ELIMINATIONS---------------")
df = df.sort_values("Eliminations")
liste_elim = list(df["Eliminations"])
affichage(liste_elim)

#accuracy normal
print("---------------PRECISION------------------")
df = df.sort_values("Accuracy")
liste_precis = list(df["Accuracy"])
affichage(liste_precis)

#degats mis normal
print("---------------DEGATS MIS------------------")
df = df.sort_values("Damage to Players")
liste_degat= list(df["Damage to Players"])
affichage(liste_degat)


#distance voyagée
print("---------------DISTANCE VOYAGEE------------------")
df = df.sort_values("Distance Traveled")
liste_dist = list(df["Distance Traveled"])
affichage(liste_dist)




```
